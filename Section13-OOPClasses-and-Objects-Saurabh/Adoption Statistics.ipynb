```python
#Import requied modules
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the GT&E data from .csv file and filter the GPN's according to the Opslist
#TimeData = pd.read_csv("GTE Extract.csv",encoding = 'ISO-8859-1')
TimeData = pd.read_csv("GTE Extract Present 09November2018 - 07December2018.csv",encoding = 'latin-1',error_bad_lines = False, sep=',')
TimeData['Transaction Week End Date'] = TimeData['Transaction Week End Date'].apply(pd.to_datetime, infer_datetime_format=True)
TimeData['Transaction Week End Date'] = TimeData['Transaction Week End Date'].apply(pd.Timestamp)
TimeData.head()
```

```python
Opslist = pd.read_csv("Tax GPN.csv",encoding = 'latin-1',error_bad_lines = False,sep=',')
Opslist.head()
```

```python
# Sub-service lines which has below 30 samples are being removed.
TimeData = TimeData.loc[(TimeData['Person SSL Name/Code'] != '"Brand,Market&Communications (0903)"') & (TimeData['Person SSL Name/Code'] != 'Other Assurance (0199)') & (TimeData['Person SSL Name/Code'] != 'CT-Valuation & Bus Modeling (0706)') & (TimeData['Person SSL Name/Code'] != 'CT-Lead Advisory (0701)') & (TimeData['Person SSL Name/Code'] != 'Risk (0304)') & (TimeData['Person SSL Name/Code'] != 'PI (0316)') & (TimeData['Person SSL Name/Code'] != 'Audit (0101)')]
TimeData.loc[(TimeData['Person SMU Name/Code'] == 'Corporate Law (0299431)'),'Person SSL Name/Code'] = 'Corporate Law (0299431)'
```

```python
FinalTimeData=pd.merge(TimeData, Opslist, on='GPN', how='inner')
FinalTimeData.head()
```

```python
FinalTimeData = FinalTimeData.drop(FinalTimeData.columns[[33]], axis=1)   #Dropping column "Name"
TimeData = FinalTimeData
TimeData.head()
```
